```python
####### TRANSFORMATIONS ######
```

```python
products['edad'] = products['edad'].str.strip()
products['edad'] = products['edad'].replace('NA',np.nan)
products['edad'] = products['edad'].str.strip().astype(float)
```

```python
products['num_antiguedad'] = products['num_antiguedad'].str.strip()
products['num_antiguedad'] = products['num_antiguedad'].replace('NA',np.nan)
products['num_antiguedad'] = products['num_antiguedad'].str.strip().astype(float)
```

```python
products.loc[products['indresi'] == 'S', 'indresi'] = 1
products.loc[products['indext'] == 'S', 'indext'] = 1
products.loc[products['xti_extra'] == 'S', 'xti_extra'] = 1
```

```python
products.loc[products['indresi'] == 'N', 'indresi'] = 0
products.loc[products['indext'] == 'N', 'indext'] = 0
products.loc[products['xti_extra'] == 'N', 'xti_extra'] = 0
```

```python
products["indresi"] = products['indresi'].astype(float)
products["indext"] = products['indext'].astype(float)
products["xti_extra"] = products['xti_extra'].astype(float)
```

```python
# quitamos la variable para no complicarnos (mas adelante ya veremos)
del products['des_canal']
del products['pais']
```

```python
#categorical data
categorical_cols = ['sexo', 'xti_empleado','tip_rel_1mes','id_segmento','mes'] 

#import pandas as pd
hot_encoded = pd.get_dummies(products, columns = categorical_cols)
hot_encoded.info()
```

```python
cols_at_end = hot_encoded.iloc[:,16:41]
cols_at_end
```

```python
# ordenamos las columnas
hot_encoded = hot_encoded[[c for c in hot_encoded if c not in cols_at_end] 
        + [c for c in cols_at_end if c in hot_encoded]]
```

```python

X = hot_encoded.iloc[:,1:45]
y = hot_encoded.iloc[:,45:46]
y.info()
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
# xg_reg = xgb.XGBRegressor(objective ='binary:logistic', colsample_bytree = 0.3, learning_rate = 0.1,
#                 max_depth = 5, alpha = 10, n_estimators = 10)


xgb_dmat = xgb.DMatrix(X_train, label=y_train)

param = {'max_depth': 1,
         'eta': 0.01,
         'objective': 'binary:logistic',
         'eval_metric': 'error',
         'nthread': 4}

model = xgb.train(param, xgb_dmat, 400)
```

```python
# xg_reg.fit(X_train,y_train)
# preds = xg_reg.predict(X_test)
preds = model.predict(xgb.DMatrix(X_test))
```
