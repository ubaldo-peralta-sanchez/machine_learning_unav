```python
####### TRANSFORMATIONS ######
```

```python
products['edad'] = products['edad'].str.strip()
products['edad'] = products['edad'].replace('NA',np.nan)
products['edad'] = products['edad'].str.strip().astype(float)
```

```python
products['num_antiguedad'] = products['num_antiguedad'].str.strip()
products['num_antiguedad'] = products['num_antiguedad'].replace('NA',np.nan)
products['num_antiguedad'] = products['num_antiguedad'].str.strip().astype(float)
```

```python
products.loc[products['indresi'] == 'S', 'indresi'] = 1
products.loc[products['indext'] == 'S', 'indext'] = 1
products.loc[products['xti_extra'] == 'S', 'xti_extra'] = 1
```

```python
products.loc[products['indresi'] == 'N', 'indresi'] = 0
products.loc[products['indext'] == 'N', 'indext'] = 0
products.loc[products['xti_extra'] == 'N', 'xti_extra'] = 0
```

```python
products["indresi"] = products['indresi'].astype(float)
products["indext"] = products['indext'].astype(float)
products["xti_extra"] = products['xti_extra'].astype(float)
```

```python
# quitamos la variable para no complicarnos (mas adelante ya veremos)
del products['des_canal']
del products['pais']
```

```python
#categorical data
categorical_cols = ['sexo', 'xti_empleado','tip_rel_1mes','id_segmento'] 

#import pandas as pd
hot_encoded = pd.get_dummies(products, columns = categorical_cols)
hot_encoded.info()
```

```python
cols_at_end = hot_encoded.iloc[:,17:42]
cols_at_end
```
